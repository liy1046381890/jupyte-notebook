```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

数据获取

```python
import pandas_datareader as pdr
from datetime import datetime
start = datetime(2017,9,1)
alibaba = pdr.get_data_yahoo("BABA", start=start)
```

```python
alibaba.head()
# alibaba.describe()
# alibaba.info()
```

趋势分析

```python
alibaba['Adj Close'].plot(legend=True)
```

```python
alibaba["High_minus_low"] = alibaba["High"] - alibaba["Low"]
```

```python
#每天之间的差异
alibaba["Daily_return"] = alibaba["Adj Close"].pct_change()
```

```python
#相邻天差异浮动
alibaba["Daily_return"].plot(figsize=(10,4), linestyle='--', marker='o')  # figsize改变图标大小
```

```python
sns.distplot(alibaba["Daily_return"].dropna(), bins=100, color='purple')
```

风险分析

```python
company = ['AAPL', 'GOOG', 'MSFT', 'AMZN', 'FB']
co_finance_df = pdr.get_data_yahoo(company, start=start)
co_finance_df.head()
```

```python
#adj close
df = co_finance_df["Adj Close"]
df.head()
```

```python
#daily return
new_df = df.pct_change()
new_df.head()
```

```python
df.plot()
```

```python
df[['AAPL', 'MSFT', 'FB']].plot()
```

```python
sns.jointplot('AAPL', 'FB', new_df, kind='scatter')
```

```python
sns.jointplot('GOOG', 'AMZN', new_df, kind='scatter')
```

```python
sns.pairplot(new_df.dropna())
```

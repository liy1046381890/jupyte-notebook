```python
#seaborn是matplotlib扩展之一
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

直方图 密度图

```python
s = Series(np.random.randn(1000))
sns.distplot(s, hist=True, kde=True, rug=True) 
```

```python
sns.kdeplot(s, shade=True)  
```

```python
s1 = Series(np.random.randn(1000))

#cbar：参数若为True，则会添加一个颜色棒(颜色帮在二元kde图像中才有)
sns.kdeplot(s, s1, shade=True, cbar=True)  
```

```python
sns.rugplot(s, axis='y', height=0.2)
```

柱状图  热力图

```python
# load a dataset from the online repository
df = sns.load_dataset('flights')
```

```python
df.year.unique()
```

```python
#透视表
df = df.pivot(index='month', columns='year', values='passengers')
df
```

```python
df.sum().plot(kind='bar')
```

```python
sns.barplot(x=df.sum().index, y=df.sum().values)
```

```python
sns.heatmap(df)
```

```python
sns.heatmap(df, annot=True, fmt='d')
```
